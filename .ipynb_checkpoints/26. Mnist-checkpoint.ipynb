```r
# 손글씨 인식
# v1 은 정답 v2~v785까지는 픽셀인데 28x28 로 되어있음 그래서 총784개  픽셀 안의 세기에 따라 숫자가 다름 
mnist <- read.csv("../Data//train_1k.csv",header=F)
head(mnist)
str(mnist)
```

```r
str(mnist[,100:150]) 
```

```r
# Data 내용 확인하기
num1 = seq(2,785,28)
num2 = seq(29,785,28)
cnt =0
for(j in 1:28){
    
    n1 = as.numeric(num1[j])
    n2 = as.numeric(num2[j])
    for(i in seq(n1,n2,1)){
        if(mnist[2,i] !=0){
            cat("*")
            
        }else{
            cat(" ")
        }
        cnt<-cnt+1
        if(cnt>27){
            cat("\n")
            cnt <-0
        }
    }
}
```

```r
# 숫자 순서대로 정렬하기
library(dplyr)
mnist <-arrange(mnist,V1)
table(mnist$V1)
mnist
```

```r
mnist$V1 <-factor(mnist$V1)
table(mnist$V1 )
```

```r
str(mnist)
```

```r
library(caret)
samp <- createDataPartition(mnist$V1, p=0.7, list=F)
data.tr <-mnist[samp,]
data.test<-mnist[-samp,]
x <- subset(data.test,select=-V1)
y <- data.test$V1
```

```r
# 인공신경망 
library(nnet)
    model.nnet =nnet(V1 ~ ., data=data.tr, size=800)
    pred=predict(model.nnet,x, type="class")
  mean(pred == y)
table(pred,y)
```

```r
# 의사결정나무 
library(party)
    model.ctree <- ctree(V1~.,data=data.tr)
    pred <- predict(model.ctree,x)
 mean(pred==y)
table(pred,y)
```

```r
# randomforest
 library(randomForest)
    model.rf <- randomForest(V1~., data = data.tr, ntree=100)
    pred <- predict(model.rf, x)
mean(pred==y)
table(pred,y)
```

```r
# svm
library(e1071)
    library(caret)

    
    model.svm <- svm(V1~.,data=data.tr)
    pred <- predict(model.svm,x, type="class")
     mean(pred==y)
    table(pred,y)
```

```r
# 로지스틱 회귀분석
library(nnet)
library(e1071)
    model <- multinom(V1 ~., data=data.tr)
    pred <- predict(model,x, type="class")
    mean(pred==y)
    table(pred,y)
```
