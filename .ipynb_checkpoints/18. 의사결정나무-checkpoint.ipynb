```r
samp <- createDataPartition(iris$Species, p=0.7, list=F) #species를 기준으로 
samp
```

```r
# 학습용
data.tr <- iris[samp,]

#테스트용
data.test <-iris[-samp,]
```

```r
# 학습시키기
model.ctree <- ctree(Species~.,data=data.tr)
```

```r
# 문제 및 정답 만들기
x <- subset (data.test, select=-Species)
y<- data.test$Species
```

```r
# 검증하기
pred <- predict(model.ctree,x)
pred
```

```r
# 예측력과 오분류표
mean(pred==y)
table(pred,y)
```
