```r
x <- 10
y <- runif(1)*100

if(x>y){
    print("x>runif()")
}
          print("-----end-----")
```

```r
x <- 10
y <- runif(1)*100

if(x>y){
    print("x>runif()")
}else{
     print("x<=runif()")
}
          print("-----end-----")
```

```r
x <- 10
y <- runif(1)*100

if(x>y){
    print("x>runif()")
}else if(x<y){
     print("x<runif()")
}else{
    print("x=runif()")
}
      
```

```r
# 홀수 짝수 구분

a <-c(8,7,6)  #if else는 벡터값을 처리못해줌 

if(a%%2==0){
   sprintf("%i은(는) 짝수",a)
}else{
    cat(a,"은(는) 홀수")
} 
```

```r
a <- c(8,7,6)

#ifelse는 벡터처리가 가능

ifelse(a %% 2 ==0, sprintf("%i : 짝수",a),sprintf("%i : 홀수",a))
```

```r
ifelse(a%%2==0,paste(a," : 짝수"),paste(a, ":홀수"))
```

```r
## 사용자로부터 data 를 입력받아 입력한 숫자가 홀수인지 짝수인지 판단하여 출력하기

num1 <- as.integer(readline("숫자를 입력하세요 :"))

# 입력한 숫자는 12이고 짝수입니다.

ifelse(num1%%2==0,sprintf("입력하신 숫자는 %i 이고 짝수입니다",num1),sprintf("입력하신 숫자는 %i 이고 홀수입니다",num1))
```

```r
score <- as.integer(readline("점수를 입력하세요 :"))
if (score>=90){
    print("A학점 입니다.")
}else if(score>=80){
    print("B학점 입니다.")
}else if (score>=70){
    print("C학점 입니다.")
}else if (score>=60){
    print("d학점 입니다.")
    
}else{
    print("F학점 입니다.")
}
```

### csv file 불러서 사용하기

```r
excel <- read.csv("../Data//excel.csv")
head(excel)
tail(excel)
str(excel)
```

```r
# q1컬럼에서 Data가 3보다 크거나 같으면 data의 제곱근을 구하고 아니면 원래 data를 출력하기

q1<-excel$q1
length(q1)

print(ifelse(q1>=3,round(sqrt(q1),4),q1))
```

```r
#q1컬럼에서 data가 3보다 크거나 같으면 data의 제곱근을 구하고 아니면 원래 data를 q6라는 컬럼을 만들어 추가하기 
excel

excel$q6 <- ifelse(excel$q1>=3,sqrt(excel$q1),excel$q1)
```
