### 한 화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10,1,100)
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해보기 

par(mfrow=c(2,2))
plot(price, type='l')
plot(price, type='o')
plot(price, type='h')
plot(price, type='s')
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해보기 

#어떻게 그리는게 좋을지 비교해볼때도 많이 사용
par(mfrow=c(2,2))
plot(price, type='o',pch=1)
plot(price,type='o', pch=2)
plot(price,type='o', pch=3, col="blue")
plot(price,type='o', pch=c(rep(15,8),20,20), cex=c(rep(1,8),3,1))
```

```r
options(warn=-1)
par(family ="AppleGothic")

#기타 그래프
x <- c(1,3,6,8,9)
y <- c(12,56,78,32,9)

plot(x,y)

# 선분 그리기 (3,3) 와 (4,4)
segments(6,78,8,32, col='red')

# 화살표 그리기
arrows(3,56,1,12,col='blue')

#사각형 그리기
rect(4,20,6,30,
     col='green',border='green',density=20)

#글자 쓰기
text(4,40,"This is 샘플", srt=50)  #srt는 각도 

# 테두리 활용하기
mtext("character",side=4, adj=0.5)  #side가 변의 위치 adj가 좌우위치

# chart의 테두리 그리기
box(lty=1, col="blue")

# x축의 추가 : y축 40 위치에 0~10까지 표시하고 색은 빨간색 
axis(1,pos=40, at=0:10, col="red")  #1이 x축 2가 y축

#y축의 추가 : x축 6위치에 20~60까지 표시하고 색은 파란색
axis(2,pos=6, at=20:60, col="blue")
```

```r
# 투표수가 높은 시군구의 후보별 득표수의 막대그래프에 시군구별 투표수 표시하기

elect<- read.csv("../Data//election_2017.csv")

str(elect)
```

```r
library(dplyr)

a<-arrange(elect,desc(pop))[1:5,-1]
a
b<-t(a[,4:6])
rownames(b) <- c("문재인","홍준표","안철수")
```

```r
options(scipen = 100)
sum<-a$moon+a$hong+a$ahn
sum[1]
```

```r
par(family = "AppleGothic")
barplot(b,
       main="투표수 높은 지역의 후보별 투표수",
    names=paste(a$광역시도,a$시군),
        ylim=c(0,500000),
        cex.names=0.6,
        col=c("skyblue","red","lightgreen"),legend.text = T 
       )

#글자 쓰기

for(i in 0:4){
text(i*1.2+0.7,sum[i+1]+10000,sum[i+1])  }
```

```r
a$문득표율 <-round(a$moon/(a$moon+a$hong+a$ahn)*100,2)
a$홍득표율 <-round(a$hong/(a$moon+a$hong+a$ahn)*100,2)
a$안득표율 <-round(a$ahn/(a$moon+a$hong+a$ahn)*100,2)
a[,7]
```

```r
par(family = "AppleGothic")
barplot(b,
       main="투표수 높은 지역의 후보별 투표수",
    names=paste(a$광역시도,a$시군),
        ylim=c(0,500000),
        cex.names=0.6,
        col=c("skyblue","red","lightgreen"),legend.text = T 
       )

#글자 쓰기

for(i in 0:4){
text(i*1.2+0.7,a$moon[i+1]/2,paste(a$문득표율[i+1],"%") )

text(i*1.2+0.7,a$moon[i+1]+a$hong[i+1]/2,paste(a$홍득표율[i+1],"%") )

text(i*1.2+0.7,a$moon[i+1]+a$hong[i+1]+a$ahn[i+1]/2,paste(a$안득표율[i+1],"%") )
}


```

```r
head(b)
head(a)
a[1,7]
par(family = "AppleGothic")
barplot(b,
       main="투표수 높은 지역의 후보별 투표수",
    names=paste(a$광역시도,a$시군),
        ylim=c(0,500000),
        cex.names=0.6,
        col=c("skyblue","red","lightgreen"),legend.text = T 
       )

# 태언오빠 방법 
for ( i in 1:5){
    for(j in 1:3){
        text(i*1.2-0.5, sum(b[1:j,i][-j] ,b[j,i]/2) , paste(round(b[j,i]/sum(b[,i])*100,2),"%") ,cex = 0.75)
    }
}
```

# 수학 함수의 시각화

y=x^2

```r
f <- function(x){
    y=x^2
    return(y)
}
```

```r
x<-1:3
f(x)
```

```r
x<-seq(-5,5,0.1)
plot(x,f(x),
    type="l"
    )
```

y = e^(-x)-4x^2 +x+3|x|-3

```r
f <- function(x){
   y = exp(-x)-4*x^2 +x+3*abs(x)-3
    return(y)
}
```

```r

plot(x,f(x),
    type="l")
```

# y= x^2 + 2x+3

```r
f <- function(x){
    y=x^2+2*x+3
    return(y)
}

```

```r
plot(x,f(x),
    type="l")

lines(x,2*x+2,
    type="l",col="red") 
```
