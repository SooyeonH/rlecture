```r
target <- read.csv("Data/newSeoul2019.csv")
data <- read.table("Data/trafficAccident_2011_2019.txt", head=T)[,-4]
data <- subset(data, 자치구별 != '합계')
colnames(data) <- c('기간','자치구별','구분',1:12)
rownames(data) <- c()

df <- data.frame(년도=1, 월=1, 자치구명=1, 발생건수=1, 사망자수=1, 부상자수=1)

for(i in 1:(nrow(data)/3)){
  년도 <- as.numeric(data[i*3, 1])
  월 <- c(1:12)
  자치구명 <- as.character(data[i*3, 2])
  발생건수 <- as.numeric(t(data[i*3-2,4:15]))
  사망자수 <- t(data[i*3-1,4:15])
  사망자수 <- as.numeric(ifelse(사망자수 == '-',0,사망자수))
  부상자수 <- as.numeric(t(data[i*3,4:15]))

  덩어리 <- data.frame(cbind(년도, 월, 자치구명, 발생건수, 사망자수, 부상자수))
  colnames(덩어리)<- c("년도","월","자치구명","발생건수","사망자수","부상자수")
  df <- rbind(df, 덩어리)
}
df <- df[-1,]
df
```
