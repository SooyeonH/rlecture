# 3D Pie Chart

```r
library(plotrix)
```

```r
sales <- c(210,110,400,550,600,130)
week <-c('월','화','수','목','금','토')
ratio <- sales/sum(sales)*100
label <- paste(week,"\n",ratio,"%")

par(family ="AppleGothic")
pie3D(sales,  #3D라고 적어주기 
   main ="영업실적 파이차트",
    labels = label,  #기본값이 레인보우여서 색지정하지 않아도 됨
      explode=0.05,
      labelcex=0.8,
      radius=0.8
   )
```
